```python
import pandas as pd

# 1. 데이터 로드 (필요한 컬럼만 선택)
prior = pd.read_csv('order_products__prior.csv', usecols=['product_id'])
products = pd.read_csv('products.csv')
aisles = pd.read_csv('aisles.csv')
departments = pd.read_csv('departments.csv')

# 2. 제품별 판매량 집계 및 상위 100개 추출
top_products = prior['product_id'].value_counts().head(100).reset_index()
top_products.columns = ['product_id', 'count']

# 3. 제품 정보(이름, 통로ID, 부서ID) 결합
top_products_info = pd.merge(top_products, products, on='product_id')
top_products_info = pd.merge(top_products_info, aisles, on='aisle_id')
top_products_info = pd.merge(top_products_info, departments, on='department_id')

# 4. 상위 제품이 많이 포함된 부서(Department) 순위
top_departments = top_products_info['department'].value_counts()

# 5. 상위 제품이 많이 포함된 통로(Aisle) 순위
top_aisles = top_products_info['aisle'].value_counts()

print("--- 상위 100개 제품이 포진한 부서 TOP 5 ---")
print(top_departments.head(5))

print("\n--- 상위 100개 제품이 포진한 통로 TOP 5 ---")
print(top_aisles.head(5))
```
