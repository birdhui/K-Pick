```python
import pandas as pd
import numpy as np
from utils import reduce_memory_usage

# 데이터 로드
orders = pd.read_csv('../data/prep/orders_mem.csv')

# 메모리 최적화
orders = reduce_memory_usage(orders, 'orders')

# 결측치 처리
# is_first_order: 이번 주문이 유저의 첫 주문인가?
orders['days_since_prior_order'] = orders['days_since_prior_order'].fillna(-1)

# 첫 주문 변수 생성
# days_since_prior: 마지막 주문 후 경과 일수
orders['is_first_order'] = (orders['days_since_prior_order'] == -1).astype(np.int8)

# 이름 정리 (가독성을 위해)
# order_dow, order_hour_of_day, order_number는 이미 존재하므로 
# 분석하기 편하게 컬럼명만 살짝 다듬거나 그대로 유지
orders.rename(columns={
    'order_hour_of_day': 'order_hour',
    'days_since_prior_order': 'days_since_prior'
}, inplace=True)

# 최종 필요한 컬럼만 선택
base_orders = orders[['user_id', 'order_id', 'eval_set', 'order_number', 
                      'order_dow', 'order_hour', 'days_since_prior', 'is_first_order']]

# float(실수) 상태인 컬럼을 다시 int16(정수)으로 강제 형변환
base_orders['days_since_prior'] = base_orders['days_since_prior'].astype(np.int16)

# 다시 한번 최적화
base_orders = reduce_memory_usage(base_orders, 'base_orders')
print(base_orders.head())

# csv 저장
base_orders.to_csv('../data/prep/base_orders.csv', index=False)
```
