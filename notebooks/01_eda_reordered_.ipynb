```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 로드
prior = pd.read_csv('../data/order_products__prior.csv')
train = pd.read_csv('../data/order_products__train.csv')

# 두 데이터 통합 (전체 상품 관계 분석용)
order_products_all = pd.concat([prior, train], axis=0)

reorder_ratio = order_products_all['reordered'].value_counts(normalize=True)
avg_order_size = train.groupby('order_id')['product_id'].count().mean()
items_per_order = order_products_all.groupby('order_id')['product_id'].count()
print(f"주문당 평균 상품 수: {avg_order_size:.2f}개")
print(f"재구매(1) 비중: {reorder_ratio[1]:.2%}")
print(f"신규구매(0) 비중: {reorder_ratio[0]:.2%}")
print(f"주문당 평균 상품 수: {items_per_order.mean():.2f}개")
print(f"가장 많이 담긴 상품 수: {items_per_order.max()}개")

# 담기 순서별 재구매율 평균 계산 (상위 30개 순서까지만)
cart_order_reorder = order_products_all.groupby('add_to_cart_order')['reordered'].mean().head(30)

plt.figure(figsize=(12, 6))
sns.lineplot(x=cart_order_reorder.index, y=cart_order_reorder.values, marker='o', color='royalblue')
plt.title('Reorder Rate by Add-to-Cart Order')
plt.xlabel('Add to Cart Order')
plt.ylabel('Reorder Rate')
plt.xticks(range(1, 31))
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 분포 시각화
plt.figure(figsize=(10, 5))
sns.histplot(items_per_order, bins=100, kde=True, color='purple')
plt.title('Distribution of Items per Order')
plt.xlabel('Number of Items')
plt.ylabel('Count')
plt.xlim(0, 50) # 가독성을 위해 50개까지만 제한
plt.show()
```
