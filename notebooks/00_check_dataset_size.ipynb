```python
import pandas as pd
import glob

# data 폴더 내의 모든 .csv 파일 목록 가져오기
file_list = glob.glob('../data/*.csv')

print("--- [데이터 파일 목록] ---")

# 각 파일의 이름과 행 개수 출력
for path in file_list:
  # 0: 부모경로, 1: 파일명 경로 -> 역슬래시 기준으로 잘라 리스트로 만든 후 마지막 요소[-1] 추출
  file_name = path.split('\\')[-1]
  df = pd.read_csv(path) # csv 파일을 데이터프레임으로 읽어오기
  
  # :,: 천 단위 구분 기호
  print(f"{file_name}: {len(df):,} 건")
```
